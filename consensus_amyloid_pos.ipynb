```python
import pandas as pd
dfclinical=pd.read_csv('pet_read.csv')
dfclinical.info()
```

```python
dfquant=pd.read_csv('allamyloid_20230104_with_compSUVR.csv')
dfquant.info()
```

```python
# rename dfclinical columns to match dfquant
dfclinical=dfclinical.rename(columns={'INDDID':'INDD','PETDate':'AMYLOIDDATE','PETTracer':'TRACER'})
```

#### match data types for merge & fix '.' to '_'

```python
dfquant.dtypes
```

```python
# dfclincical['INDD']: float64
# dfquant['INDD']: object
```

```python
dfclinical["INDD"]=dfclinical["INDD"].astype(str)
for index, row in dfclinical.iterrows():
    if len(row['INDD']) == 9:       
        dfclinical.at[index,'INDD']=row['INDD'].replace('.', '_')
    else:
        dfclinical.at[index,'INDD']=row['INDD'].replace('.0', '')
```

```python
for index, row in dfclinical.iterrows():
    datelist=row['AMYLOIDDATE'].split('/')
    print(datelist)
    dateint=int(datelist[2] + datelist[0] + datelist[1])
    print(dateint)
    dfclinical.at[index,'AMYLOIDDATE']=dateint
```

#### remove dupes from pet_read dfclinical

```python
dfclinical[dfclinical.duplicated(subset=['INDD','AMYLOIDDATE'],keep=False)]
```

```python
dfclin_nodupes=dfclinical.drop_duplicates(subset=['INDD','AMYLOIDDATE'],keep='first')
dfclin_nodupes.info()
```

#### merge clinical reading and quant data

```python
#merging on INDD, AMYLOIDDATE; AgeatPET, TRACER (for now) and Clinical_Read are new columns
dfcombine2 = pd.merge(dfclin_nodupes,dfquant,how='outer',on=['INDD', 'AMYLOIDDATE'],suffixes=('_clinical', '_quant'))
dfcombine2.info()
```

```python
dfcombine2.head()
```

#### reformatting tracer columns to create match

```python
##word match--use FBB and Florbetapir for clinical tracer
for index, row in dfcombine2.iterrows():
    if pd.isna(dfcombine2.loc[index,'TRACER_clinical']):
        continue
    elif row['TRACER_clinical']=='Florbetaben (amyloid)':
        dfcombine2.at[index,'TRACER_clinical']='FBB'
    elif row['TRACER_clinical'].lower()=='florbetapir (amyloid)':
        dfcombine2.at[index,'TRACER_clinical']='Florbetapir'
    else:
        print(row['TRACER_clinical'])
```

```python
for index, row in dfcombine2.iterrows():
    if row['TRACER_clinical'] == row['TRACER_quant'] or pd.isna(dfcombine2.loc[index,'TRACER_quant']):
        dfcombine2.at[index,'TRACER']=row['TRACER_clinical']
        print('Clinical field and quant match, or quant field is nan--using clinical field for Tracer column')
    elif pd.isna(dfcombine2.loc[index,'TRACER_clinical']):
        print(f"Clinical field NaN")
        dfcombine2.at[index,'TRACER']=row['TRACER_quant']
    else:
        print(f"Clinical: {row['TRACER_clinical']}, Quant: {row['TRACER_quant']}")
        dfcombine2.at[index,'TRACER']=row['TRACER_clinical']
```

```python
dfcombine2.head(10)
```

#### Create Quant_read column

```python
for index, row in dfcombine2.iterrows():
    if row['TRACER']=='Florbetapir':
        if row['compSUVR'] >= 1.11:
#             print("Florbetapir, over 1.11 = row['Quant_read']='Positive'")
            dfcombine2.at[index,'Quant_read']='Positive'
        elif pd.isna(dfcombine2.loc[index,'compSUVR']):
            print('no quant data')
            continue
        else:
            dfcombine2.at[index,'Quant_read']='Negative'
    elif row['TRACER']=='FBB':
        if row['compSUVR'] >= 1.08:
#             print("FBB, over 1.08 = Positive")
            dfcombine2.at[index,'Quant_read']='Positive'
        elif pd.isna(dfcombine2.loc[index,'compSUVR']):
            print('no quant data')
            continue
        else:
#             print("Negative")
            dfcombine2.at[index,'Quant_read']='Negative'
```

#### check for clinical read on sessions with comments on quant data

```python
for index, row in dfcombine2.iterrows():
    if pd.notnull(dfcombine2.loc[index,'Comments']):
        print(f"ID {row['INDD']}, PetDate {row['AMYLOIDDATE']}: {row['Comments']}")
        if pd.notnull(dfcombine2.loc[index,'Clinical_Read']):
            print(f'\tClinical data exists for this session')
        else:
            print(f'\tNo clinical data')
```

#### create consensus between clinical and quant reads

```python
dfcombine_copy=dfcombine2.copy()
#copy just in case
```

```python
dfcombine2['Amyloid_consensus']=''
dfcombine2['Amyloid_Consensus_Comments']=''
```

```python
dfcombine2.rename(columns={"Comments": "Processing_Comments"},inplace=True)
```

```python
agree=0
noclinicaldata=0
noquantdata=0
clinical_inconclusive=0
disagree=0

for index, row in dfcombine2.iterrows():
    if row['Quant_read'] == row['Clinical_Read']:  ##Match
        dfcombine2.at[index,'Amyloid_consensus']=row['Clinical_Read']
        agree+=1
        
    elif pd.isna(dfcombine2.loc[index,'Clinical_Read']):  ##No clinical data, use quant
        noclinicaldata+=1  
        if pd.notnull(dfcombine2.loc[index,"Processing_Comments"]): ##there is a comment about processing
            dfcombine2.at[index,'Amyloid_consensus']='Inconclusive'
            dfcombine2.at[index,'Amyloid_Consensus_Comments']='Amyloid consensus inconclusive until processing issue resolved, no clinical read'
        else:
            dfcombine2.at[index,'Amyloid_consensus']=row['Quant_read'] 
            dfcombine2.at[index,'Amyloid_Consensus_Comments']='Amyloid consensus used quantitative result, Clinical_Read does not exist'
            
    elif pd.isna(dfcombine2.loc[index,'compSUVR']):  ##No Quantitative data, use clinical
        dfcombine2.at[index,'Amyloid_consensus']=row['Clinical_Read']
        dfcombine2.at[index,'Amyloid_Consensus_Comments']='Amyloid consensus used clinical read, no quantitative results'
        noquantdata+=1

    elif row['Clinical_Read'] == 'Inconclusive':  ##Inconclusive clinical, use quant
        clinical_inconclusive +=1
        dfcombine2.at[index,'Amyloid_consensus']=row['Quant_read']
        dfcombine2.at[index,'Amyloid_Consensus_Comments']='Amyloid consensus used quantitative result, Clinical_read inconclusive'
       
    else:
        dfcombine2.at[index,'Amyloid_consensus']=row['Clinical_Read'] ##Quant and Clin don't match
        dfcombine2.at[index,'Amyloid_Consensus_Comments']='Amyloid consensus used clinical read result, quantitative result does not match'
        disagree+=1        
        
print(f'Agree: {agree}, No clinical data: {noclinicaldata}, No Quantitative data: {noquantdata}, Clinical Inconclusive: {clinical_inconclusive} Disagree: {disagree}')
```
